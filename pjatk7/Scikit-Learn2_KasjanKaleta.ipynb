```python
# ============================================================
# ZADANIE 1: Wczytanie i wstępne przygotowanie danych
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# ── 1. Wczytanie danych ──────────────────────────────────────
df = pd.read_csv("marketing_campaign.csv", sep=None, engine="python")

print("=== SHAPE ===")
print(df.shape)

print("\n=== INFO ===")
df.info()

print("\n=== STATYSTYKI OPISOWE ===")
display(df.describe(include="all"))

# ── 2. Brakujące wartości ────────────────────────────────────
print("\n=== BRAKUJĄCE WARTOŚCI ===")
missing = df.isnull().sum()
print(missing[missing > 0])

# Uzupełniamy brakujące wartości Income medianą
df["Income"] = df["Income"].fillna(df["Income"].median())

print("\nBrakujące wartości po uzupełnieniu:", df.isnull().sum().sum())

# ── 3. Kolumna CustomerFrom → liczba dni jako klient ─────────
df["CustomerFrom"] = pd.to_datetime(df["CustomerFrom"], dayfirst=True)
reference_date = df["CustomerFrom"].max()
df["CustomerDays"] = (reference_date - df["CustomerFrom"]).dt.days
df.drop(columns=["CustomerFrom"], inplace=True)

print("\n=== CustomerDays ===")
print(df["CustomerDays"].describe())

# ── 4. Kodowanie zmiennych nienumerycznych ───────────────────
cat_cols = df.select_dtypes(include="object").columns.tolist()
print("\n=== Kolumny kategoryczne ===", cat_cols)
for col in cat_cols:
    print(f"  {col}: {df[col].unique()}")

# -- Education: zmienna PORZĄDKOWA
education_order = {
    "Basic": 0,
    "2n Cycle": 1,
    "Graduation": 2,
    "Master": 3,
    "PhD": 4
}
df["Education"] = df["Education"].map(education_order)

# -- MaritalStatus: zmienna NOMINALNA → one-hot
df["MaritalStatus"] = df["MaritalStatus"].replace({
    "Alone": "Single",
    "Absurd": "Single",
    "YOLO": "Single"
})
df = pd.get_dummies(df, columns=["MaritalStatus"], drop_first=False, dtype=int)

print("\n=== Kolumny po kodowaniu ===")
print(df.columns.tolist())

# ── 5. Wartości odstające i skalowanie ───────────────────────

# Wizualizacja boxplotów
num_cols = df.select_dtypes(include=np.number).columns.tolist()

fig, axes = plt.subplots(nrows=(len(num_cols) // 4) + 1, ncols=4, figsize=(20, 20))
axes = axes.flatten()
for i, col in enumerate(num_cols):
    axes[i].boxplot(df[col].dropna())
    axes[i].set_title(col, fontsize=9)
for j in range(i + 1, len(axes)):
    axes[j].set_visible(False)
plt.suptitle("Boxploty zmiennych numerycznych", fontsize=14)
plt.tight_layout()
plt.show()

# Usuwanie outlierów metodą IQR (factor=3.0)
def remove_outliers_iqr(data, cols, factor=3.0):
    mask = pd.Series([True] * len(data), index=data.index)
    for col in cols:
        Q1 = data[col].quantile(0.25)
        Q3 = data[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - factor * IQR
        upper = Q3 + factor * IQR
        mask &= data[col].between(lower, upper)
    return data[mask]

outlier_cols = ["Income", "YearOfBirth",
                "AmountWines", "AmountMeat"]
print(f"\nRozmiar przed usunięciem outlierów: {df.shape}")
df_clean = remove_outliers_iqr(df, outlier_cols, factor=3.0)
print(f"Rozmiar po usunięciu outlierów:    {df_clean.shape}")

# Skalowanie
scaler = StandardScaler()
X_scaled_array = scaler.fit_transform(df_clean)
X_scaled = pd.DataFrame(X_scaled_array, columns=df_clean.columns, index=df_clean.index)

print("\n=== Przeskalowany zbiór – próbka ===")
display(X_scaled.head())

print("\n✅ Zadanie 1 zakończone.")
print(f"   Wymiary końcowego zbioru: {X_scaled.shape}")
```

```python
# ============================================================
# ZADANIE 2: Grupowanie – podejście pierwsze
# ============================================================

from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.metrics import silhouette_samples
from scipy.cluster.hierarchy import dendrogram, linkage
import matplotlib.cm as cm
import warnings
warnings.filterwarnings("ignore")

# Używamy X_scaled z Zadania 1
X = X_scaled.values

# ── 1. DENDROGRAM (grupowanie hierarchiczne) ─────────────────
# Na próbce 200 punktów – pełny zbiór byłby nieczytelny
np.random.seed(42)
sample_idx = np.random.choice(len(X), size=200, replace=False)
X_sample = X[sample_idx]

linked = linkage(X_sample, method="ward")

plt.figure(figsize=(18, 6))
dendrogram(
    linked,
    truncate_mode="lastp",
    p=30,
    leaf_rotation=90,
    leaf_font_size=10,
    show_contracted=True
)
plt.title("Dendrogram (Ward, próbka 200 pkt)", fontsize=14)
plt.xlabel("Liczba punktów w węźle")
plt.ylabel("Odległość (Ward)")
plt.axhline(y=25, color="red", linestyle="--", label="sugerowane cięcie")
plt.legend()
plt.tight_layout()
plt.show()

# ── 2. METRYKI vs LICZBA KLASTRÓW (k-means) ──────────────────
k_range = range(2, 11)
inertia_list = []
silhouette_list = []
db_list = []
ch_list = []

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X)
    inertia_list.append(km.inertia_)
    silhouette_list.append(silhouette_score(X, labels))
    db_list.append(davies_bouldin_score(X, labels))
    ch_list.append(calinski_harabasz_score(X, labels))

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

axes[0, 0].plot(k_range, inertia_list, "bo-")
axes[0, 0].set_title("Metoda łokcia – Inercja (K-Means)")
axes[0, 0].set_xlabel("Liczba klastrów k")
axes[0, 0].set_ylabel("Inercja")

axes[0, 1].plot(k_range, silhouette_list, "go-")
axes[0, 1].set_title("Współczynnik sylwetki (K-Means)")
axes[0, 1].set_xlabel("Liczba klastrów k")
axes[0, 1].set_ylabel("Silhouette score")

axes[1, 0].plot(k_range, db_list, "ro-")
axes[1, 0].set_title("Indeks Daviesa-Bouldina (K-Means)")
axes[1, 0].set_xlabel("Liczba klastrów k")
axes[1, 0].set_ylabel("Davies-Bouldin (niższy = lepszy)")

axes[1, 1].plot(k_range, ch_list, "mo-")
axes[1, 1].set_title("Indeks Calińskiego-Harabasza (K-Means)")
axes[1, 1].set_xlabel("Liczba klastrów k")
axes[1, 1].set_ylabel("Calinski-Harabasz (wyższy = lepszy)")

plt.suptitle("Metryki jakości klasteryzacji vs liczba klastrów", fontsize=14)
plt.tight_layout()
plt.show()

# Wydruk tabelaryczny
print("\n=== TABELA METRYK (K-Means) ===")
print(f"{'k':>4} | {'Inercja':>12} | {'Silhouette':>10} | {'Davies-Bouldin':>14} | {'Calinski-H':>12}")
print("-" * 62)
for i, k in enumerate(k_range):
    print(f"{k:>4} | {inertia_list[i]:>12.1f} | {silhouette_list[i]:>10.4f} | {db_list[i]:>14.4f} | {ch_list[i]:>12.2f}")

# ── 3. WYKRESY SYLWETKI dla k=2..5 ───────────────────────────
def plot_silhouette(X, k, ax):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(X)
    sil_vals = silhouette_samples(X, labels)
    avg = silhouette_score(X, labels)

    y_lower = 10
    colors = cm.nipy_spectral(np.linspace(0, 1, k))
    for i in range(k):
        ith_sil = np.sort(sil_vals[labels == i])
        size = ith_sil.shape[0]
        y_upper = y_lower + size
        ax.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_sil,
                         facecolor=colors[i], alpha=0.7)
        ax.text(-0.05, y_lower + 0.5 * size, str(i))
        y_lower = y_upper + 10

    ax.axvline(x=avg, color="red", linestyle="--")
    ax.set_title(f"k={k}  (avg={avg:.3f})")
    ax.set_xlabel("Współczynnik sylwetki")
    ax.set_ylabel("Klaster")
    ax.set_xlim([-0.1, 1])

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
for idx, k in enumerate([2, 3, 4, 5]):
    plot_silhouette(X, k, axes[idx // 2][idx % 2])
plt.suptitle("Wykresy sylwetki – K-Means", fontsize=14)
plt.tight_layout()
plt.show()

# ── 4. NAJLEPSZA LICZBA KLASTRÓW → finalny model ─────────────
# Na podstawie wykresów wybieramy k
best_k = silhouette_list.index(max(silhouette_list)) + 2
print(f"\n→ Najlepsza k wg Silhouette: {best_k}")

km_final = KMeans(n_clusters=best_k, random_state=42, n_init=10)
km_labels = km_final.fit_predict(X)

print(f"\n=== K-MEANS (k={best_k}) – METRYKI ===")
print(f"  Silhouette:       {silhouette_score(X, km_labels):.4f}")
print(f"  Davies-Bouldin:   {davies_bouldin_score(X, km_labels):.4f}")
print(f"  Calinski-Harabasz:{calinski_harabasz_score(X, km_labels):.2f}")
print(f"  Liczba punktów w klastrach: {np.bincount(km_labels)}")

# ── 5. GRUPOWANIE HIERARCHICZNE (Agglomerative) ───────────────
print("\n=== GRUPOWANIE HIERARCHICZNE ===")
sil_agg = []
for k in k_range:
    agg = AgglomerativeClustering(n_clusters=k, linkage="ward")
    labels_agg = agg.fit_predict(X)
    sil_agg.append(silhouette_score(X, labels_agg))

best_k_agg = sil_agg.index(max(sil_agg)) + 2
print(f"→ Najlepsza k wg Silhouette (Agglomerative): {best_k_agg}")

agg_final = AgglomerativeClustering(n_clusters=best_k_agg, linkage="ward")
agg_labels = agg_final.fit_predict(X)

print(f"\n=== AGGLOMERATIVE (k={best_k_agg}) – METRYKI ===")
print(f"  Silhouette:       {silhouette_score(X, agg_labels):.4f}")
print(f"  Davies-Bouldin:   {davies_bouldin_score(X, agg_labels):.4f}")
print(f"  Calinski-Harabasz:{calinski_harabasz_score(X, agg_labels):.2f}")
print(f"  Liczba punktów w klastrach: {np.bincount(agg_labels)}")

print("\n✅ Zadanie 2 zakończone.")
```

"""
DENDROGRAM (Ward):
- Największy skok odległości (~45) widoczny jest między 2 a 3 gałęziami,
  co sugeruje naturalny podział na 2 grupy.
- Po cięciu na poziomie ~25 widoczne są 3 gałęzie (niebieski klaster
  i dwie zielone), co sugeruje również k=3 jako alternatywę.
- Dendrogram nie wskazuje wyraźnie na k>3 jako naturalny podział.

METRYKI K-MEANS:
- Silhouette: najwyższy dla k=2 (0.2215), potem spada – wyraźny sygnał,
  że dane naturalnie dzielą się na 2 grupy.
- Davies-Bouldin: najniższy (lepszy) dla k=2 (1.83), potwierdza k=2.
- Calinski-Harabasz: monotonicznie malejący – preferuje k=2.
- Inercja (łokieć): brak wyraźnego "łokcia", krzywa opada płynnie.

WYKRESY SYLWETKI:
- k=2: avg=0.222 – najlepszy wynik; oba klastry mają zbliżone rozmiary
  (1307 i 900), choć wiele punktów ma niskie wartości sylwetki (blisko 0),
  co oznacza że granica między grupami nie jest ostra.
- k=3: avg=0.155 – wyraźny spadek, jeden klaster (szary) zdominowany
  przez punkty z ujemną sylwetką.
- k=4,5: dalszy spadek jakości podziału.

GRUPOWANIE HIERARCHICZNE (Ward, k=4):
- Agglomerative wybrało k=4 (Silhouette=0.157), ale jest to gorszy wynik
  niż k=2 w K-Means (0.222).
- Podział na klastry: 534 / 944 / 501 / 228 – nierównomierny,
  klaster z 228 punktami może być grupą specyficznych klientów.

PODSUMOWANIE:
- Wszystkie metryki i dendrogram zgodnie wskazują k=2 jako optymalną
  liczbę klastrów dla tego zbioru danych w przestrzeni 23 wymiarów.
- Ogólnie niskie wartości Silhouette (<0.25) sugerują duże nakładanie
  się grup – co jest typowe dla danych wysokowymiarowych (klątwa wymiarowości).
- W kolejnych zadaniach (inżynieria cech + PCA) spodziewamy się poprawy
  jakości klasteryzacji.
"""

```python
# ============================================================
# ZADANIE 3: Grupowanie – DBSCAN
# ============================================================

from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import warnings
warnings.filterwarnings("ignore")

# Używamy X_scaled z Zadania 1
X = X_scaled.values

# Na podstawie dendrogramu – najbardziej prawdopodobna liczba klastrów
TARGET_CLUSTERS = 2

# ── a) Wykres k-odległości (metoda łokcia dla epsilon) ───────
# MinPts typowo = 2 * liczba_wymiarów, ale sprawdzimy kilka wartości
# Standardowo k = 4 lub 5 dla wykresu k-odległości

k_neighbors = 5
nbrs = NearestNeighbors(n_neighbors=k_neighbors).fit(X)
distances, _ = nbrs.kneighbors(X)
k_distances = np.sort(distances[:, k_neighbors - 1])[::-1]  # k-ta odległość, malejąco

plt.figure(figsize=(10, 5))
plt.plot(k_distances)
plt.title(f"Wykres k-odległości (k={k_neighbors}) – szukamy 'kolana'")
plt.xlabel("Punkty posortowane malejąco")
plt.ylabel(f"{k_neighbors}-ta najbliższa odległość")
plt.grid(True)
plt.tight_layout()
plt.show()

# ── b) Siatka parametrów eps x MinPts ────────────────────────
# Na podstawie wykresu k-odległości – szacunkowy zakres eps

eps_range = np.arange(3.5, 5.5, 0.1)
minpts_range = range(2, 9)

results = []

print("Szukam optymalnych parametrów DBSCAN...")
for eps in eps_range:
    for minpts in minpts_range:
        db = DBSCAN(eps=eps, min_samples=minpts)
        labels = db.fit_predict(X)

        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        n_noise = np.sum(labels == -1)

        # Zapisujemy tylko jeśli liczba klastrów == TARGET_CLUSTERS
        if n_clusters == TARGET_CLUSTERS and n_clusters > 1:
            sil = silhouette_score(X[labels != -1], labels[labels != -1]) \
                  if len(set(labels[labels != -1])) > 1 else -1
            results.append({
                "eps": round(eps, 2),
                "minpts": minpts,
                "n_clusters": n_clusters,
                "n_noise": n_noise,
                "silhouette": sil
            })

if results:
    results_df = pd.DataFrame(results).sort_values("silhouette", ascending=False)
    print(f"\nZnaleziono {len(results)} kombinacji z {TARGET_CLUSTERS} klastrami.")
    print("\nTop 10 kombinacji:")
    print(results_df.head(10).to_string(index=False))

    # ── c) Najlepsza kombinacja ───────────────────────────────
    best = results_df.iloc[0]
    best_eps = best["eps"]
    best_minpts = int(best["minpts"])

    print(f"\n→ Najlepsza kombinacja: eps={best_eps}, MinPts={best_minpts}")

    db_final = DBSCAN(eps=best_eps, min_samples=best_minpts)
    best_labels = db_final.fit_predict(X)

    n_clusters_final = len(set(best_labels)) - (1 if -1 in best_labels else 0)
    n_noise_final = np.sum(best_labels == -1)
    labels_unique = set(best_labels) - {-1}

    print(f"\n=== DBSCAN – WYNIKI ===")
    print(f"  Liczba klastrów:          {n_clusters_final}")
    print(f"  Liczba punktów szumu:     {n_noise_final}")
    print(f"  Etykiety klastrów:        {sorted(labels_unique)}")
    print(f"  Punkty w klastrach:")
    for lbl in sorted(labels_unique):
        print(f"    Klaster {lbl}: {np.sum(best_labels == lbl)} punktów")

    # Metryki tylko na punktach nie-szumowych
    mask = best_labels != -1
    sil_final  = silhouette_score(X[mask], best_labels[mask])
    db_idx     = davies_bouldin_score(X[mask], best_labels[mask])
    ch_idx     = calinski_harabasz_score(X[mask], best_labels[mask])

    print(f"\n=== METRYKI JAKOŚCI (bez punktów szumu) ===")
    print(f"  Silhouette:        {sil_final:.4f}")
    print(f"  Davies-Bouldin:    {db_idx:.4f}")
    print(f"  Calinski-Harabasz: {ch_idx:.2f}")

    # ── Wykres sylwetki DBSCAN ────────────────────────────────
    fig, ax = plt.subplots(figsize=(8, 5))
    sil_vals = silhouette_samples(X[mask], best_labels[mask])
    labels_clean = best_labels[mask]
    unique_labels = sorted(set(labels_clean))
    colors = cm.nipy_spectral(np.linspace(0, 1, len(unique_labels)))

    y_lower = 10
    for i, lbl in enumerate(unique_labels):
        ith_sil = np.sort(sil_vals[labels_clean == lbl])
        size = len(ith_sil)
        y_upper = y_lower + size
        ax.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_sil,
                         facecolor=colors[i], alpha=0.7)
        ax.text(-0.05, y_lower + 0.5 * size, str(lbl))
        y_lower = y_upper + 10

    ax.axvline(x=sil_final, color="red", linestyle="--",
               label=f"avg = {sil_final:.3f}")
    ax.set_title(f"Wykres sylwetki – DBSCAN (eps={best_eps}, MinPts={best_minpts})")
    ax.set_xlabel("Współczynnik sylwetki")
    ax.set_ylabel("Klaster")
    ax.set_xlim([-0.1, 1])
    ax.legend()
    plt.tight_layout()
    plt.show()

else:
    print(f"\n⚠️  Nie znaleziono kombinacji dającej dokładnie {TARGET_CLUSTERS} klastry.")
    print("Spróbuj zmienić zakres eps_range lub TARGET_CLUSTERS.")

print("\n✅ Zadanie 3 zakończone.")
```

"""
WYKRES K-ODLEGŁOŚCI:
- Brak wyraźnego "kolana" – krzywa opada płynnie bez gwałtownego zagięcia.
- Jest to typowe dla danych wysokowymiarowych (23 wymiary) – klątwa
  wymiarowości sprawia, że odległości między punktami są do siebie podobne.
- Szacowany zakres eps: 4.0–5.5 (delikatne wygięcie przy pierwszych ~50 pkt).

WYNIKI DBSCAN (eps=5.4, MinPts=8):
- Znaleziono 2 klastry (zgodnie z sugestią dendrogramu).
- Klaster 0: 2117 punktów (96.5% danych) – główna masa klientów.
- Klaster 1: 76 punktów (3.5%) – mała, wyraźnie odstająca grupa klientów.
- Szum: tylko 14 punktów (0.6%) – bardzo mało, co oznacza że eps jest
  dość duże i algorytm "wciągnął" prawie wszystko do klastrów.

METRYKI JAKOŚCI:
- Silhouette: 0.2506 – LEPSZY niż K-Means k=2 (0.2215)!
- Davies-Bouldin: 1.5063 – LEPSZY niż K-Means k=2 (1.8271), niższy = lepszy.
- Calinski-Harabasz: 113.43 – gorszy niż K-Means (625.05), ale CH faworyzuje
  równomiernie rozłożone klastry, a tu mamy bardzo nierówny podział.

WYKRES SYLWETKI:
- Klaster 1 (76 pkt) ma wysokie, jednorodne wartości sylwetki – jest to
  dobrze wyodrębniona, zwarta grupa klientów.
- Klaster 0 (2117 pkt) ma szerokie, zróżnicowane wartości (wiele bliskich 0)
  – wewnętrznie jest bardzo niejednorodny, co sugeruje że kryje w sobie
  kilka podgrup.
- Silhouette avg=0.251 jest wyższy niż dla K-Means, ale wynika głównie
  z bardzo dobrze wyodrębnionego małego klastra 1.

PODSUMOWANIE:
- DBSCAN zidentyfikował małą grupę "wyjątkowych" klientów (klaster 1, 76 os.)
  wyraźnie różniących się od reszty – prawdopodobnie klienci o bardzo
  wysokich wydatkach lub specyficznym profilu.
- Słabe "kolano" i nierównomierny podział potwierdzają że dane w tej
  przestrzeni 23 wymiarów są trudne do klasteryzacji – redukcja wymiarów
  (Zadania 4 i 5) powinna znacząco poprawić wyniki.
"""

```python
# ============================================================
# ZADANIE 4: Grupowanie – podejście trzecie (inżynieria cech)
# ============================================================

# ── 1. Nowe cechy na bazie df_clean ─────────────────────────
df4 = df_clean.copy()

# Spent – suma wydatków we wszystkich kategoriach
amount_cols = ["AmountWines", "AmountFruits", "AmountMeat",
               "AmountFish", "AmountSweets", "AmountGoldProds"]
df4["Spent"] = df4[amount_cols].sum(axis=1)

# Children – łączna liczba potomstwa
df4["Children"] = df4["KidsHome"] + df4["TeensHome"]

# FamilySize – wielkość gospodarstwa domowego
# MaritalStatus_Married + MaritalStatus_Together = para (2 os.), reszta = 1 os.
df4["FamilySize"] = df4["Children"] + \
    np.where(df4["MaritalStatus_Married"] + df4["MaritalStatus_Together"] > 0, 2, 1)

# IsParent – czy ma dzieci (0/1)
df4["IsParent"] = (df4["Children"] > 0).astype(int)

print("=== Nowe cechy (próbka) ===")
print(df4[["Spent", "Children", "FamilySize", "IsParent"]].describe())

# ── 2. Usuwamy kolumny źródłowe nowych cech ──────────────────
cols_to_drop = amount_cols + ["KidsHome", "TeensHome",
                               "MaritalStatus_Married", "MaritalStatus_Together",
                               "MaritalStatus_Divorced", "MaritalStatus_Single",
                               "MaritalStatus_Widow"]
df4_reduced = df4.drop(columns=cols_to_drop)

print(f"\nShape zbioru pierwotnego (X_scaled): {X_scaled.shape}")
print(f"Shape zbioru rozszerzonego (df4_reduced): {df4_reduced.shape}")
print(f"Kolumny: {df4_reduced.columns.tolist()}")

# ── 3. Skalowanie nowego zbioru ───────────────────────────────
scaler4 = StandardScaler()
X4_scaled_array = scaler4.fit_transform(df4_reduced)
X4 = pd.DataFrame(X4_scaled_array, columns=df4_reduced.columns, index=df4_reduced.index)

print(f"\nShape X4 (przeskalowany): {X4.shape}")

# ── 4. K-MEANS na nowym zbiorze ───────────────────────────────
print("\n" + "="*50)
print("K-MEANS – nowy zbiór cech")
print("="*50)

k_range = range(2, 11)
sil4_km = []
for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    lbl = km.fit_predict(X4.values)
    sil4_km.append(silhouette_score(X4.values, lbl))

best_k4 = sil4_km.index(max(sil4_km)) + 2
print(f"→ Najlepsza k wg Silhouette: {best_k4}")

km4 = KMeans(n_clusters=best_k4, random_state=42, n_init=10)
km4_labels = km4.fit_predict(X4.values)

print(f"\n=== K-MEANS (k={best_k4}) – METRYKI ===")
print(f"  Silhouette:        {silhouette_score(X4.values, km4_labels):.4f}")
print(f"  Davies-Bouldin:    {davies_bouldin_score(X4.values, km4_labels):.4f}")
print(f"  Calinski-Harabasz: {calinski_harabasz_score(X4.values, km4_labels):.2f}")
print(f"  Liczba punktów:    {np.bincount(km4_labels)}")

# Wykres sylwetki K-Means
fig, ax = plt.subplots(figsize=(8, 5))
plot_silhouette(X4.values, best_k4, ax)
ax.set_title(f"Sylwetka – K-Means k={best_k4} (inżynieria cech)")
plt.tight_layout()
plt.show()

# ── 5. DBSCAN na nowym zbiorze ────────────────────────────────
print("\n" + "="*50)
print("DBSCAN – nowy zbiór cech")
print("="*50)

# Wykres k-odległości dla nowego zbioru
nbrs4 = NearestNeighbors(n_neighbors=5).fit(X4.values)
dist4, _ = nbrs4.kneighbors(X4.values)
k_dist4 = np.sort(dist4[:, 4])[::-1]

plt.figure(figsize=(10, 4))
plt.plot(k_dist4)
plt.title("Wykres k-odległości (k=5) – nowy zbiór cech")
plt.xlabel("Punkty posortowane malejąco")
plt.ylabel("5-ta najbliższa odległość")
plt.grid(True)
plt.tight_layout()
plt.show()

# Siatka parametrów DBSCAN
eps4_range = np.arange(1.5, 4.0, 0.1)
results4 = []

for eps in eps4_range:
    for minpts in range(2, 9):
        db = DBSCAN(eps=eps, min_samples=minpts)
        labels = db.fit_predict(X4.values)
        n_cl = len(set(labels)) - (1 if -1 in labels else 0)
        if n_cl == best_k4 and n_cl > 1:
            mask = labels != -1
            if mask.sum() > 0 and len(set(labels[mask])) > 1:
                sil = silhouette_score(X4.values[mask], labels[mask])
                results4.append({
                    "eps": round(eps, 2), "minpts": minpts,
                    "n_clusters": n_cl,
                    "n_noise": int(np.sum(labels == -1)),
                    "silhouette": sil
                })

if results4:
    res4_df = pd.DataFrame(results4).sort_values("silhouette", ascending=False)
    best4 = res4_df.iloc[0]
    print(f"→ Najlepsza kombinacja: eps={best4['eps']}, MinPts={int(best4['minpts'])}")

    db4 = DBSCAN(eps=best4["eps"], min_samples=int(best4["minpts"]))
    db4_labels = db4.fit_predict(X4.values)

    n_cl4 = len(set(db4_labels)) - (1 if -1 in db4_labels else 0)
    n_noise4 = int(np.sum(db4_labels == -1))
    mask4 = db4_labels != -1

    print(f"\n=== DBSCAN (nowe cechy) – WYNIKI ===")
    print(f"  Liczba klastrów:      {n_cl4}")
    print(f"  Liczba punktów szumu: {n_noise4}")
    for lbl in sorted(set(db4_labels) - {-1}):
        print(f"  Klaster {lbl}: {np.sum(db4_labels == lbl)} punktów")

    print(f"\n=== METRYKI JAKOŚCI (bez szumu) ===")
    print(f"  Silhouette:        {silhouette_score(X4.values[mask4], db4_labels[mask4]):.4f}")
    print(f"  Davies-Bouldin:    {davies_bouldin_score(X4.values[mask4], db4_labels[mask4]):.4f}")
    print(f"  Calinski-Harabasz: {calinski_harabasz_score(X4.values[mask4], db4_labels[mask4]):.2f}")

    # Wykres sylwetki DBSCAN
    fig, ax = plt.subplots(figsize=(8, 5))
    sil_vals4 = silhouette_samples(X4.values[mask4], db4_labels[mask4])
    labels_cl4 = db4_labels[mask4]
    colors4 = cm.nipy_spectral(np.linspace(0, 1, n_cl4))
    y_lower = 10
    for i, lbl in enumerate(sorted(set(labels_cl4))):
        ith = np.sort(sil_vals4[labels_cl4 == lbl])
        y_upper = y_lower + len(ith)
        ax.fill_betweenx(np.arange(y_lower, y_upper), 0, ith,
                         facecolor=colors4[i], alpha=0.7)
        ax.text(-0.05, y_lower + 0.5 * len(ith), str(lbl))
        y_lower = y_upper + 10
    avg4 = silhouette_score(X4.values[mask4], db4_labels[mask4])
    ax.axvline(x=avg4, color="red", linestyle="--", label=f"avg={avg4:.3f}")
    ax.set_title(f"Sylwetka – DBSCAN (nowe cechy, eps={best4['eps']}, MinPts={int(best4['minpts'])})")
    ax.set_xlabel("Współczynnik sylwetki")
    ax.set_ylabel("Klaster")
    ax.set_xlim([-0.1, 1])
    ax.legend()
    plt.tight_layout()
    plt.show()
else:
    print(f"⚠️  Brak kombinacji z {best_k4} klastrami. Dostosuj eps4_range.")

print("\n✅ Zadanie 4 zakończone.")
```

```python
# ============================================================
# ZADANIE 5: Grupowanie – podejście czwarte (PCA)
# ============================================================

from sklearn.decomposition import PCA

# ── 1. Standaryzacja pełnej ramki df4 (z nowymi cechami) ─────
# WAŻNE: standaryzujemy df4 (pełny, z nowymi cechami), nie df4_reduced
scaler5 = StandardScaler()
X4_full = df4.drop(columns=cols_to_drop, errors="ignore")

# Upewniamy się że używamy df4_reduced (już bez źródłowych kolumn)
X5_std = scaler5.fit_transform(df4_reduced)

print(f"Zbiór do PCA – shape: {X5_std.shape}")
print(f"Liczba cech: {df4_reduced.shape[1]}")

# ── 2. Iteracja po liczbie składowych – skumulowana wariancja ─
n_features = df4_reduced.shape[1]
cumvar = []

for n in range(1, n_features + 1):
    pca_tmp = PCA(n_components=n, random_state=42)
    pca_tmp.fit(X5_std)
    cumvar.append(np.sum(pca_tmp.explained_variance_ratio_))

# ── 3. Wykres skumulowanej wariancji ─────────────────────────
plt.figure(figsize=(9, 5))
plt.plot(range(1, n_features + 1), cumvar, "bo-")
plt.axhline(y=0.70, color="red", linestyle="--", label="70% wariancji")
plt.xlabel("Liczba składowych głównych")
plt.ylabel("Skumulowana wariancja")
plt.title("PCA – skumulowana wariancja wyjaśniona")
plt.xticks(range(1, n_features + 1))
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ── 4. Wybór liczby składowych ≈ 70% wariancji ───────────────
n_components = next(i + 1 for i, v in enumerate(cumvar) if v >= 0.70)
print(f"\n→ Liczba składowych dla ≥70% wariancji: {n_components}")
print(f"  Skumulowana wariancja: {cumvar[n_components-1]:.4f}")

# Szczegóły każdej składowej
pca_full = PCA(random_state=42)
pca_full.fit(X5_std)
print("\n=== Wariancja wyjaśniana przez kolejne składowe ===")
for i, (ev, cv) in enumerate(zip(pca_full.explained_variance_ratio_, cumvar)):
    marker = " ←" if i + 1 == n_components else ""
    print(f"  PC{i+1:2d}: {ev:.4f}  (skum: {cv:.4f}){marker}")

# ── 5. Nowa macierz zmiennych po PCA ─────────────────────────
pca = PCA(n_components=n_components, random_state=42)
X_pca = pca.fit_transform(X5_std)

print(f"\nShape po PCA: {X_pca.shape}")

# ── 6. K-MEANS na danych PCA ──────────────────────────────────
print("\n" + "="*50)
print("K-MEANS – dane po PCA")
print("="*50)

sil_pca_km = []
for k in range(2, 11):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    lbl = km.fit_predict(X_pca)
    sil_pca_km.append(silhouette_score(X_pca, lbl))

best_k_pca = sil_pca_km.index(max(sil_pca_km)) + 2
print(f"→ Najlepsza k wg Silhouette: {best_k_pca}")

km_pca = KMeans(n_clusters=best_k_pca, random_state=42, n_init=10)
km_pca_labels = km_pca.fit_predict(X_pca)

print(f"\n=== K-MEANS PCA (k={best_k_pca}) – METRYKI ===")
print(f"  Silhouette:        {silhouette_score(X_pca, km_pca_labels):.4f}")
print(f"  Davies-Bouldin:    {davies_bouldin_score(X_pca, km_pca_labels):.4f}")
print(f"  Calinski-Harabasz: {calinski_harabasz_score(X_pca, km_pca_labels):.2f}")
print(f"  Liczba punktów:    {np.bincount(km_pca_labels)}")

# Wykres sylwetki
fig, ax = plt.subplots(figsize=(8, 5))
plot_silhouette(X_pca, best_k_pca, ax)
ax.set_title(f"Sylwetka – K-Means PCA k={best_k_pca}")
plt.tight_layout()
plt.show()

# ── 7. DBSCAN na danych PCA ───────────────────────────────────
print("\n" + "="*50)
print("DBSCAN – dane po PCA")
print("="*50)

# Wykres k-odległości
nbrs_pca = NearestNeighbors(n_neighbors=5).fit(X_pca)
dist_pca, _ = nbrs_pca.kneighbors(X_pca)
k_dist_pca = np.sort(dist_pca[:, 4])[::-1]

plt.figure(figsize=(10, 4))
plt.plot(k_dist_pca)
plt.title("Wykres k-odległości (k=5) – dane po PCA")
plt.xlabel("Punkty posortowane malejąco")
plt.ylabel("5-ta najbliższa odległość")
plt.grid(True)
plt.tight_layout()
plt.show()

# Siatka parametrów
eps_pca_range = np.arange(0.5, 3.0, 0.1)   # mniejsze eps po PCA!
results_pca = []

for eps in eps_pca_range:
    for minpts in range(2, 9):
        db = DBSCAN(eps=eps, min_samples=minpts)
        labels = db.fit_predict(X_pca)
        n_cl = len(set(labels)) - (1 if -1 in labels else 0)
        if n_cl == best_k_pca and n_cl > 1:
            mask = labels != -1
            if mask.sum() > 0 and len(set(labels[mask])) > 1:
                sil = silhouette_score(X_pca[mask], labels[mask])
                results_pca.append({
                    "eps": round(eps, 2), "minpts": minpts,
                    "n_clusters": n_cl,
                    "n_noise": int(np.sum(labels == -1)),
                    "silhouette": sil
                })

if results_pca:
    res_pca_df = pd.DataFrame(results_pca).sort_values("silhouette", ascending=False)
    best_pca = res_pca_df.iloc[0]
    print(f"→ Najlepsza kombinacja: eps={best_pca['eps']}, MinPts={int(best_pca['minpts'])}")
    print(f"\nTop 5:\n{res_pca_df.head().to_string(index=False)}")

    db_pca = DBSCAN(eps=best_pca["eps"], min_samples=int(best_pca["minpts"]))
    db_pca_labels = db_pca.fit_predict(X_pca)

    n_cl_pca = len(set(db_pca_labels)) - (1 if -1 in db_pca_labels else 0)
    n_noise_pca = int(np.sum(db_pca_labels == -1))
    mask_pca = db_pca_labels != -1

    print(f"\n=== DBSCAN PCA – WYNIKI ===")
    print(f"  Liczba klastrów:      {n_cl_pca}")
    print(f"  Liczba punktów szumu: {n_noise_pca}")
    for lbl in sorted(set(db_pca_labels) - {-1}):
        print(f"  Klaster {lbl}: {np.sum(db_pca_labels == lbl)} punktów")

    sil_p  = silhouette_score(X_pca[mask_pca], db_pca_labels[mask_pca])
    db_p   = davies_bouldin_score(X_pca[mask_pca], db_pca_labels[mask_pca])
    ch_p   = calinski_harabasz_score(X_pca[mask_pca], db_pca_labels[mask_pca])

    print(f"\n=== METRYKI JAKOŚCI (bez szumu) ===")
    print(f"  Silhouette:        {sil_p:.4f}")
    print(f"  Davies-Bouldin:    {db_p:.4f}")
    print(f"  Calinski-Harabasz: {ch_p:.2f}")

    # Wykres sylwetki DBSCAN PCA
    fig, ax = plt.subplots(figsize=(8, 5))
    sil_vals_p = silhouette_samples(X_pca[mask_pca], db_pca_labels[mask_pca])
    labels_p = db_pca_labels[mask_pca]
    colors_p = cm.nipy_spectral(np.linspace(0, 1, n_cl_pca))
    y_lower = 10
    for i, lbl in enumerate(sorted(set(labels_p))):
        ith = np.sort(sil_vals_p[labels_p == lbl])
        y_upper = y_lower + len(ith)
        ax.fill_betweenx(np.arange(y_lower, y_upper), 0, ith,
                         facecolor=colors_p[i], alpha=0.7)
        ax.text(-0.05, y_lower + 0.5 * len(ith), str(lbl))
        y_lower = y_upper + 10
    ax.axvline(x=sil_p, color="red", linestyle="--", label=f"avg={sil_p:.3f}")
    ax.set_title(f"Sylwetka – DBSCAN PCA (eps={best_pca['eps']}, MinPts={int(best_pca['minpts'])})")
    ax.set_xlabel("Współczynnik sylwetki")
    ax.set_ylabel("Klaster")
    ax.set_xlim([-0.1, 1])
    ax.legend()
    plt.tight_layout()
    plt.show()
else:
    print(f"⚠️  Brak kombinacji z {best_k_pca} klastrami. Dostosuj eps_pca_range.")

# ── 8. TABELA PORÓWNAWCZA wszystkich zadań ────────────────────
print("\n" + "="*60)
print("TABELA PORÓWNAWCZA WSZYSTKICH METOD")
print("="*60)
print(f"{'Metoda':<35} | {'Silhouette':>10} | {'DB':>8} | {'CH':>10}")
print("-"*60)
print(f"{'KMeans k=2 (oryg. 23 cechy)':<35} | {0.2215:>10.4f} | {1.8271:>8.4f} | {625.05:>10.2f}")
print(f"{'Agglomerative k=4 (oryg.)':<35} | {0.1567:>10.4f} | {2.3071:>8.4f} | {295.21:>10.2f}")
print(f"{'DBSCAN (oryg. 23 cechy)':<35} | {0.2506:>10.4f} | {1.5063:>8.4f} | {113.43:>10.2f}")
print(f"{'KMeans k=2 (inż. cech, 14)':<35} | {0.2540:>10.4f} | {1.5793:>8.4f} | {805.61:>10.2f}")
print(f"{'DBSCAN (inż. cech, 14)':<35} | {0.2721:>10.4f} | {1.5298:>8.4f} | {657.62:>10.2f}")

print("\n✅ Zadanie 5 zakończone.")
```
